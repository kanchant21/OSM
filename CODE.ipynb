```python
import pandas as pd
```

```python
import regex as re
```

```python
# Define the Unicode for Marathi Language
marathi_pattern = re.compile('[\u0900-\u097F]+')
```

```python
type(marathi_pattern)
```

```python
# Define a function to check if the rows of name column has any marathi charcaters
def extract_marathi(row):
  name = row['name']
  matches = marathi_pattern.findall(name)
  if matches:
    row['name:mr'] = ''.join(matches)
    row['name'] = marathi_pattern.sub('', name)
    return row
```

```python
# Apply the changes
dataframe1 = dataframe1.apply(extract_marathi, axis=1)
```

```python
#Transfering marathi values from name:en to name:mr 
# Renaming the name:en tag to name_en
C1_dataframe.rename(columns = {'name:en':'name_en'}, inplace = True)
```

```python
# Define a function to check if the rows of name:en column has any marathi charcaters
def extract_marathin(row):
  name_en = row['name_en']
  matches = marathi_pattern.findall(name_en)
  if matches:
    row['name:mr'] = ''.join(matches)
    row['name_en'] = marathi_pattern.sub('', name_en)
    return row
```

```python
# Apply the changes
C1_dataframe = C1_dataframe.apply(extract_marathin, axis=1)
```

```python
import string
```

```python
# Define a function to check the rows of name:en column if there are any characters other than english and if there are transfer them to new column Non_english so it can be further checked if there are any Marathi Characters
def is_english(s):
  return all(c in string.printable for c in s)
```

```python
dataframe1['Non_english'] = ''
```

```python
for index, row in dataframe1.iterrows():
  if not is_english(str(row['name:en'])):
    non_english = ''.join([c for c in str(row['name:en']) if not is_english(c)])
    dataframe1.at[index, 'Non_english'] = non_english
```

```python
def extract_marathi_new(row):
  Non_english = row['Non_english']
  matches = marathi_pattern.findall(Non_english)
  if matches:
    row['name:mr'] = ''.join(matches)
    row['Non_english'] = marathi_pattern.sub('', Non_english)
    return row
```
